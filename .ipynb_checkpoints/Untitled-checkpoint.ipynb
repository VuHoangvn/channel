```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []
z = []
with open("output/mode/converge_no-dem1_r25_1.out") as f:
    for line in f:
        arr = line.split()
        x.append(float(arr[0]))
        y.append(float(arr[1]))
        z.append(float(arr[2]))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zline = np.linspace(30, 48, 1000)
xline = 0.5 + abs(np.sin(zline))
yline = 60 *abs(np.cos(zline))
# ax.plot3D(xline, yline, zline, 'gray')
ax.scatter3D(x, y, z, c=z, cmap='Reds');
ax.plot(x,yz)
```
